```python
import numpy as np
from numpy import linalg
```

```python
A = np.array([[3, 5, 1, 0], [5, 2, 0, 1]]) # Matriz dos coeficientes das inequações
b = np.array([[15], [10]]) # "Respostas" das inequações
C = np.array([[-5], [-3], [0], [0]]) # Função objetivo
Xb = np.array([2, 3])
Xn = np.array([0, 1])
```

```python
def simplex(A, b, C, Xb, Xn):
    B = np.array([i[Xb]for i in A])
    N = np.array([i[Xn]for i in A])
    Cb = C[Xb].flatten()
    Cn = C[Xn].flatten()


    for count in range(10):
        x_b = linalg.inv(B) @ b
        z_b = Cb.T @ x_b

        # Encontrando o Xn que entrará na base
        z_j, pos_in = 0, 0
        for i in range(0, len(Xn)-1):
            a_i = [n[i] for n in N]
            z_i = Cb @ linalg.inv(B) @ a_i  - (Cn[i])

            if z_i > 0 and  z_i > z_j:
                z_j, pos_in = z_i, i

        # Parada caso a solução ideal seja encontrada
        if z_j <= 0:
            break

        # Encontrando o Xb que saíra da base
        a_i = [i[pos_in] for i in A]
        y_i = linalg.inv(B) @ a_i

        pos_out = np.argmin(x_b.flatten()/y_i)

        # Adicionando e removendo variáveis da base
        Xb = np.append(Xb, Xn[pos_in])
        Xn = np.append(Xn, Xb[pos_out])
        Xb = np.delete(Xb, pos_out)
        Xn = np.delete(Xn, pos_in)

        # Atualizando os valores dos vetores/matrizes
        B = np.array([i[Xb]for i in A])
        N = np.array([i[Xn]for i in A])
        Cb = C[Xb].flatten()
        Cn = C[Xn].flatten()

    print("B: ", B, "\n")
    print("N: ", N, "\n")
    print("Cb: ", Cb, "\n")
    print("Cn: ", Cn, "\n")
```

```python
simplex(A, b, C, Xb, Xn)
```
