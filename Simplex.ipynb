```python
A = np.array([[1, 7, 3, 7, 1, 0, 0], [3, -1, 1, 2, 0, 1, 0], [2, 3, -1, 1, 0, 0, 1]]) # Matriz dos coeficientes das inequações
b = np.array([[46], [8], [10]]) # "Respostas" das inequações
C = np.array([[3], [6], [-2], [4], [0], [0], [0]]) # Função objetivo
Xb = np.array([4, 5, 6])
Xn = np.array([0, 1, 2, 3])
```

```python
def simplex(A, b, C, Xb, Xn):
    B = np.array([i[Xb]for i in A])
    N = np.array([i[Xn]for i in A])
    Cb = C[Xb].flatten()
    Cn = C[Xn].flatten()


    for count in range(10):
        B_inv = linalg.inv(B)
        x_b = B_inv @ b
        z_b = Cb.T @ x_b

        # Encontrando o Xn que entrará na base
        z_j, pos_in = 0, 0
        for i in range(0, len(Xn)-1):
            a_i = A[:, Xn[i]]
            z_i = Cb @ B_inv @ a_i  - (Cn[i])

            if z_i > 0 and  z_i > z_j:
                z_j, pos_in = z_i, i

        # Parada caso a solução ideal seja encontrada
        if z_j <= 0:
            break

        # Encontrando o Xb que saíra da base
        a_i = A[:, Xn[pos_in]]
        y_i = B_inv @ a_i
        x_b = x_b.flatten()

        ratios = np.array([x_b[i]/y_i[i] if y_i[i] > 0 else np.inf for i in range(len(y_i))])
        pos_out = np.argmin(ratios)

        # Adicionando e removendo variáveis da base
        leaving = Xb[pos_out]
        entering = Xn[pos_in]

        Xb[pos_out] = entering
        Xn[pos_in] = leaving

        # Atualizando os valores dos vetores/matrizes
        B = np.array([i[Xb]for i in A])
        N = np.array([i[Xn]for i in A])
        Cb = C[Xb].flatten()
        Cn = C[Xn].flatten()

    print("Xb: \n", Xb+1, "\n")
    print("Xn: \n", Xn+1, "\n")
```

```python
simplex(A, b, C, Xb, Xn)
```
